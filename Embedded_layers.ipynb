```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['This is a good boy',
      'My name is khan',
      'This is veru useful channel',
      'Data science is love ',
      'Your videos are very helpful',
      'my father name is mr  Ashok kumar gehlot']
```

```python
sent
```

```python
#vocabulary size
voc_size=10000
```

```python
onehot_repn=[one_hot(words,voc_size)for words in sent]
```

```python
onehot_repn
```

```python
#word embeding represantation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
```

```python
embedded_docs=pad_sequences(onehot_repn,padding='pre',maxlen=sent_length)
```

```python
embedded_docs
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```

```python
print(model.predict(embedded_docs[0]))
```
